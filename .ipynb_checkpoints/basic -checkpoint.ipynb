#### 元组和列表
元组可以看作轻量型列表，都属于有序，支持双向索引、切片操作，支持运算符+、*和in
元组不可变，列表可变
元组中元素的值不能直接修改，也不能为元组增加或删除元素
元组访问速度比列表快、开销小，适合对于仅便利的对象
元组可作为字典的键但是列表不可以，包含列表的元组也不可以

#### 序列解包
本质是对多个变量同时进行赋值，要求等号左侧变量数量和等号右侧值的数量一致，用于字典时默认是对字典的键进行操作，需要对“键：值”这一元素进行操作需要使用字典的items()方法明确说明，若需要对“值”进行操作，需要使用values()方法指定

```python
x,y,z=1,2,3
print(x,y,z)
x,y,z=(False,3.5,'exp')
print(x,y,z)
x,y,z=[1,2,3]
print(x,y,z)
x,y=y,x
print(x,y)
x,y,z=map(int,'123')
print(x,y,z)
data={'a':97,'b':98}
x,y=data.values()
print(x,y)
```

### 字典
字典中的键不能是列表、集合、字典或其他可变类型（可以是所有不可变类型），包含可变类型的不可变类型也不可以作为键，且不可重复，“键：值”这样的元素是无序的

```python
data=dict(name='张三',age=18,sex='M')
print(data)
data=dict.fromkeys([1,2,3,4])#指定数据为“键”
print(data)
data=dict(zip('7890',[97,98,99,100]))
print(data)
data={ch:ord(ch) for ch in 'sikhdiah'}#字典推导式
print(data)
```

##### 字典元素的访问
可以以键为下标返回对应的值，不存在则抛出异常，get()方法用于获取指定键的元素
可以用for循环遍历，默认是遍历键，需要遍历元素则必须使用items()方法，若要遍历值则使用values()
使用len(),max(),min(),sum(),sorted(),enumerate(),map(),filter()等内置函数以及成员测试运算符in对字典对象操作时，也遵循同样的约定

```python
data=dict(name='张三',age=18,sex='M')
print(data)
print(data.get('age'))
print(data.get('address'))
print(list(data))
print(list(data.values()))
print(list(data.items()))
for key,value in data.items():
    print(key,value,sep='\t')
```

##### 字典元素的修改添加与访问
以指定键为下标为元素赋值时，若已存在键则修改值，若不存在则添加一个新元素
使用字典对象的update()方法可以将另一个字典的元素一次性全部添加到当前字典中，若存在相同的键则以另一个字典中的值为准
使用字典对象的pop()方法可以删除指定键对应的元素，同时返回对应的值，popitem()方法用于删除并返回一个包含两个元素的元组，也可以用del删除指定的键对应的元素

```python
sock={'IP':'127.0.0.1','port':80}
sock['port']=8080#修改
sock['protocol']='TCP'#添加
print(sock)
sock={'IP':'127.0.0.1','port':80}
sock.update(sock={'IP':'129.168.9.62','protocol':'TCP'})
print(sock)
sock={'IP': '129.168.9.62', 'port': 8080, 'protocol': 'TCP'}
print(sock.pop('IP'))
print(sock.popitem())
del sock['port']
print(sock)
```

### 集合
无序可变的容器对象，一个集合内的每个元素唯一不可重复，只能包含不可变类型数据，不支持下标访问，不支持random中的choice()函数从集合随机选取元素，但支持random中的sample()函数随机选取部分元素；set()可以将列表元组字符串range对象转变为集合，也可以用集合推导式生成集合，del可以删除整个集合

```python
data={30,40,50}
data.add(20)#增加20
data.add(50)#已存在50忽略
data.update({40,60})#已存在40忽略
print(data)
data={30,40,50}
data.remove(30)#删除30若不存在则异常
data.discard(30)#若不存在则忽略
print(data.pop())
print(data)
```

#### 3-5生成包含1000个随机数字字符的字符串,然后统计每个数字出现的次数

```python
from random import choice#从指定分布中随机选择k个允许重复的元素,sample函数不允许重复
from string import digits
z=''.join(choice(digits) for i in range(1000))
print(z)
result={}
for ch in z:
    result[ch]=result.get(ch,0)+1
for digit, fre in sorted(result.items()):
    print(digit,fre,sep=':')
```

### 文件操作

##### open(file,mode='r',buffering=-1,encoding=None,errors=None,newline=None,closefd=True,oprner=None)
##### file:文件名称,文件不在当前文件夹时要带上路径
##### mode:指定打开文件后的处理方式--'r'为默认读模式可忽略,若文件不存在则抛出异常;'w'写模式,文件已存在会先清空不存在再创建;'x'写模式文件若存在则抛出异常;'a'追加模式;'b'二进制模式,此时不允许指定encoding参数;'t'文本模式;'+'读写模式可与其他模式组合
##### encoding指定文件解码编码方式,只适用于文本模式
##### PS:操作完文件一定关闭文件

##### close()将缓存区内容写入文件,同时关闭文件释放文件对象
##### read([size])从文本文件中读取并返回size个字符,或者从二进制文件中读取并返回size个字节,默认为全部内容
##### readline()从文本文件中读取并返回一行内容
##### readlines()从文本文件中返回包含每行内容的列表
##### seek(cookie,whence=0,/)定位文件指针,将之移动到相对于whence的偏移cookie的位置,whence为0时代表文件头1为当前位置2为文件尾,文本文件为2时cookie必须是0,二进制文件可以是负数;默认为0
##### write(s)将s的内容写入文件其中s的内容必须和文件类型一致
##### writelines(s)将列表s中的所有字符串写入文本文件但是不会自动换行(需要自己保证)
##### 其中read() write() readline()这些方法读写文件内容时表示当前位置的文件指针会自动向后移动

```python
#with open(filename,mode,encoding) as fp:
    #通过文件对象fp读写文件内容

#合并两个.txt文件内容,多行内容交替写入结果文件, 若一个文件内容较少则把另一个文件的剩余内容写入结果文件尾部
def mergeTxt(txtFiles):
    with open('result.txt','w') as fp:
        with open(txtFiles[0]) as fp1,open(txtFiles[1]) as fp2:
            while True:
                #交替读取文件1文件2中的行写入结果文件
                line1=fp1.readline()
                if line1:
                    fp.write(line1)
                else:
                    #文件1结束
                    flag=False
                    break
                line2=fp2.readline()
                if line2:
                    fp.write(line2)
                else:
                    flag=False
                    break
            #获取尚未结束的文件对象
            fp3=fp1 if flag else fp2
            for line in fp3:
                fp.write(line)
txtFiles=['1.txt','2.txt']
mergeTxt(txtFiles)
```

##### 5-3编写程序,模拟生成某饭店自2020年1月1日开始,连续100天试营业期间的营业额数据,并写入csv文件;文件共两列,第一列为日期,第二列为营业额,文件第一行为表头或字段名称.假设该饭店第一天营业额基数为500元,每天增加5元,初次之外每天还会随机增加5到50元不等

```python
from csv import reader, writer
from random import randrange
from datetime import date, timedelta

fn='data.csv'
with open(fn,'w') as fp:
    wr=writer(fp)#创建csv
    wr.writerow(['日期','      销量'])#写入表头
    startDate=date(2020,1,1)#第一天日期
    for i in range(100):
        amount=500+i*5+randrange(5,50)
        wr.writerow([str(startDate),amount])
        startDate=startDate+timedelta(days=1)
with open(fn) as fp:
    for line in reader(fp):
        if line:
            print(*line)
```
