```python
import torch, torchvision
print(torch.version.cuda)      # 应输出 12.1
print(torch.cuda.is_available())  # True → GPU 可用
```

```python
x = torch.randn(2, 3, 224, 224).cuda()
model = torchvision.models.resnet50(weights='IMAGENET1K_V1').cuda()
with torch.no_grad():
    out = model(x)
print(out.shape)   # 能跑即 OK
```
